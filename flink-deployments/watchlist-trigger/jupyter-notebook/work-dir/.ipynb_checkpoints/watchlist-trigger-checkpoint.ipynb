## User Count - added a product to their cart but not yet placed the order

```python
from delta import *
from pyspark.sql import SparkSession
from pyspark import SparkConf
from pathlib import Path
from pyspark.sql import functions as F
import plotly.express as px
from pyspark.sql.functions import col
import warnings
import plotly.io as pio
pio.renderers.default = "iframe"
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', category=UserWarning, message="Using an existing Spark session; only runtime SQL configurations will take effect.")

conf = SparkConf().setAppName('Low Inventory Alert Count')\
        .set("spark.jars.packages", "io.delta:delta-core_2.12:2.3.0")\
        .set("spark.sql.catalog.spark_catalog","org.apache.spark.sql.delta.catalog.DeltaCatalog")\
        .set("spark.sql.extensions","io.delta.sql.DeltaSparkSessionExtension")\
        .set("log4j.logger.org.apache.hadoop.util.NativeCodeLoader", "ERROR")\
        .set("log4j.logger.org.apache.spark.internal.config.native-code-path", "ERROR")

#builder = spark.builder.appName("MyApp") \
#    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
#    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")


spark = SparkSession.builder.config(conf=conf).getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

#spark = configure_spark_with_delta_pip(builder).getOrCreate()

delta_table_path = "/opt/spark/delta-tables/watchlist-trigger-alerts"

# Read data from the Delta table
#df = spark.read.format("delta").load("/opt/spark/delta-tables/low_stock_alert")
#print(f"# of Low Stock alerts: {df.count()}")

if Path(delta_table_path, "_delta_log").exists():
    df = spark.read.format("delta").load(delta_table_path)
    # Convert total_sale_price to a standard numeric format
    #df = df.withColumn("total_sale_pice", col("total_sale_pice").cast("decimal(20, 8)"))
    
    # Group by `productName` and count distinct `userId` for each product
    user_count_df = df.groupBy("productName").agg(F.countDistinct("userId").alias("userCount"))

    # Convert to Pandas DataFrame if not already
    user_count_pd = user_count_df.toPandas()

    # Sort and group by `userCount`
    #user_count_pd = user_count_pd.sort_values(by="userCount", ascending=False)

    # Create the grouped bar chart
    fig = px.bar(
        user_count_pd,
        x="productName",
        y="userCount",
        color="userCount",  # Color by userCount to distinguish the groups
        title="Grouped Bar Chart of Products by User Count",
        labels={"userCount": "User Count", "productName": "Product Name"},
        color_continuous_scale=px.colors.sequential.Viridis
    )

    # Update layout for readability
    fig.update_layout(
        xaxis_tickangle=-45,
        xaxis_title="Product Name",
        yaxis_title="User Count",
        legend_title="User Count",
        bargap=0.2
    )

    # Display the chart
    fig.show()
else:
    print("No Data available currently")
```
