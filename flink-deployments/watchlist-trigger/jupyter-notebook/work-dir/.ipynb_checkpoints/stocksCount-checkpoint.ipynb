### User based stock purchases and sales

```python
%%sql
SELECT
  userid,
  SUM(
    CASE
      WHEN side = 'BUY' THEN 1
      ELSE 0
    END
 ) AS num_purchases,
  SUM(
    CASE
      WHEN side = 'SELL' THEN 1
      ELSE 0
    END
  ) AS num_sells
FROM
  trades
GROUP BY
  userid
ORDER BY
  userid
```

### Total Stocks Purchased and Sold per User

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

total_trades = """
SELECT userid,
       side,
       COUNT(*) AS count
FROM trades
GROUP BY userid, side
ORDER BY userid, side
"""
# Execute the query
result = spark.sql(total_trades)
pandas_df = result.toPandas()

# Set the style for the plot
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x="userid", y="count", hue="side", data=pandas_df, palette="muted")

# Add titles and labels
plt.title("Stocks Purchased and Sold")
plt.xlabel("User ID")
plt.ylabel("# of Trades")
plt.legend(title='Trade Side', loc='upper right')

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Top Selling Stocks using Seaborn

```python
top_selling_stocks = """
SELECT symbol,
       SUM(quantity) AS total_volume
FROM trades
WHERE side = 'BUY'
GROUP BY symbol
ORDER BY total_volume DESC
"""

# Execute the query
result = spark.sql(top_selling_stocks)
pandas_df = result.toPandas()

plt.figure(figsize=(12, 6))
sns.barplot(x="total_volume", y="symbol", hue="total_volume", data=pandas_df, palette="Blues_d")

# Add titles and labels
plt.title("Stock Volumes")
plt.xlabel("Volume")
plt.ylabel("Ticker Symbol")

# Show the plot
plt.tight_layout()
plt.show()
```

### Top Selling Stocks using Plotly

```python
import plotly.offline as pyo
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "iframe"

fig = px.bar(pandas_df, 
             x="total_volume", 
             y="symbol", 
             orientation='h', 
             title="Stock Volumes",
             labels={"total_volume": "Total Quantity Bought", "symbol": "Ticker Symbol"},
             color="total_volume",  # Color based on the quantity sold
             color_continuous_scale=px.colors.sequential.Blues)

fig.show()
```

### Top Stocks being Sold by users - Line Chart

```python
import plotly.express as px

top_withdrawn_stocks = """
SELECT symbol,
       SUM(quantity) AS total_volume
FROM trades
WHERE side = 'SELL'
GROUP BY symbol
ORDER BY total_volume DESC
"""

# Execute the query
result = spark.sql(top_withdrawn_stocks)
top_withdrawn_stocks_df = result.toPandas()

# Line Chart Configuration
fig_line = px.line(top_withdrawn_stocks_df, 
                   x='symbol', 
                   y='total_volume', 
                   title='Top Stocks Sold by Users - Line Chart',
                   labels={'total_volume': 'Total Volume', 'symbol': 'Ticker Symbol'},
                   markers=True)  # Markers can help in visual clarity

fig_line.show()
```

### Top Stocks being Sold by users - Sunburst Chart

```python
import plotly.offline as pyo
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "iframe"

# Adding a dummy hierarchy for the sunburst chart
top_withdrawn_stocks_df['category'] = 'Top Stocks'  # This will be the top level

# Sunburst Chart Configuration
fig_sunburst = px.sunburst(top_withdrawn_stocks_df, 
                           path=['category', 'symbol'], 
                           values='total_volume', 
                           title='Top Stocks Sold by Users - Sunburst Chart',
                           labels={'total_volume': 'Total Volume', 'symbol': 'Ticker Symbol'})

fig_sunburst.show()
```

### User based Trades on a given day

```python
import plotly.offline as pyo
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "iframe"
import warnings
warnings.filterwarnings('ignore')
# Suppress specific warning
warnings.filterwarnings('ignore', category=UserWarning, message="Not authorized: invalid_client: client_id is missing")

user_trades_query = """
SELECT 
    DATE(FROM_UNIXTIME(timestamp / 1000)) AS trade_date, 
    side as txn_type, 
    userid,
    COUNT(*) AS count
FROM 
    `lkc-wkr20w`.`tableflow-stock-trades-poc`
GROUP BY 
    trade_date, 
    txn_type,
    userid
ORDER BY 
    trade_date,
    txn_type,
    userid;
"""

user_trades_df = spark.sql(user_trades_query).toPandas()

# Create the grouped bar chart
fig = px.bar(
    user_trades_df,
    x='userid',
    y='count',
    color='txn_type',
    barmode='group',
    title='Buy and Sell Transactions by User on Specific Dates',
    labels={'count': 'Number of Transactions', 'userid': 'User'},
    facet_col='trade_date',  # Create a separate subplot for each trade date
)

fig.show()
```

### Locate Users

```python
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
# Suppress specific warning
warnings.filterwarnings('ignore', category=UserWarning, message="Not authorized: invalid_client: client_id is missing")
import plotly.io as pio
pio.renderers.default = "iframe"

joined_df = spark.sql("""
    SELECT 
        c.id AS customer_id,
        c.name AS customer_name,
        c.zip AS customer_zip,
        c.state AS customer_state,
        t.userid AS trade_userid,
        c.email AS email,
        COUNT(CASE WHEN t.side = 'BUY' THEN 1 END) AS buy_count,
        COUNT(CASE WHEN t.side = 'SELL' THEN 1 END) AS sell_count,
        SUM(t.quantity) AS total_quantity,  
        SUM(t.price * t.quantity) AS total_value
    FROM customers c
    JOIN trades t ON c.name = t.userid
    GROUP BY 
        c.id, c.name, c.zip, t.userid , c.state, c.email
    ORDER BY t.userid ASC
""")

joined_pandas_df = joined_df.toPandas()

zip_to_location = {
    '90210': (34.0901, -118.4065),  # Beverly Hills
    '10001': (40.7128, -74.0060),   # New York City
    '73301': (30.2672, -97.7431),   # Austin
    '33101': (25.7617, -80.1918),   # Miami
    '60601': (41.8781, -87.6298),   # Chicago
    '44101': (41.4995, -81.6954),   # Cleaveland
    '48188': (42.3080, -83.4810),   # Canton
    '19101': (39.9526, -75.1652),   # Philadelphia
    '98101': (47.6062, -122.3321)   # Seatle
    
}

joined_pandas_df['latitude'] = joined_pandas_df['customer_zip'].map(lambda x: zip_to_location.get(x, (None, None))[0])
joined_pandas_df['longitude'] = joined_pandas_df['customer_zip'].map(lambda x: zip_to_location.get(x, (None, None))[1])

# Filter out any rows with missing lat/lon
filtered_df = joined_pandas_df.dropna(subset=['latitude', 'longitude'])

filtered_df['buy_count'] = filtered_df['buy_count'].astype(str)
filtered_df['sell_count'] = filtered_df['sell_count'].astype(str)
filtered_df['total_value'] = filtered_df['total_value'].astype(str)

filtered_df['hover_text'] = 'BUY COUNT' + ' - ' + filtered_df['buy_count'] + ',\nSELL COUNT' + ' - ' + filtered_df['sell_count'] + ' , \nTOTAL VALUE ' + ' - $' + filtered_df['total_value'] + ' , \nContact: '+ filtered_df['email'] 
filtered_df['user'] = filtered_df['customer_name'] + '(' + filtered_df['customer_state'] + ')'
fig = px.scatter_geo(filtered_df,
                     lat='latitude',
                     lon='longitude',
                     text='user',
                     hover_name='hover_text',
                     title='User Locations by Zip Code',
                     scope='usa') 

fig.update_layout(
    width=1200,   # Set the width (in pixels)
    height=800    # Set the height (in pixels)
)

fig.show()
```

### Locate States where trading is most Active

```python
import plotly.express as px
import plotly.subplots as sp
import plotly.io as pio
pio.renderers.default = "iframe"

import warnings
warnings.filterwarnings('ignore')
# Suppress specific warning
warnings.filterwarnings('ignore', category=UserWarning, message="Not authorized: invalid_client: client_id is missing")

states_df = spark.sql("""
    SELECT 
    DATE(FROM_UNIXTIME(t.timestamp / 1000)) AS trade_date,
    c.state AS customer_state,
    c.zip AS customer_zip,
    SUM(t.quantity) AS total_purchased_quantity
FROM 
    customers c
JOIN 
    trades t ON c.name = t.userid
WHERE
    t.side = 'BUY'
GROUP BY 
    trade_date, c.state, c.zip
ORDER BY 
    trade_date, total_purchased_quantity DESC
""")

joined_pandas_df = states_df.toPandas()

zip_to_location = {
    '90210': (34.0901, -118.4065),  # Beverly Hills
    '10001': (40.7128, -74.0060),   # New York City
    '73301': (30.2672, -97.7431),   # Austin
    '33101': (25.7617, -80.1918),   # Miami
    '60601': (41.8781, -87.6298),   # Chicago
    '44101': (41.4995, -81.6954),   # Cleaveland
    '48188': (42.3080, -83.4810),   # Canton
    '19101': (39.9526, -75.1652),   # Philadelphia
    '98101': (47.6062, -122.3321)   # Seatle
    
}

joined_pandas_df['latitude'] = joined_pandas_df['customer_zip'].map(lambda x: zip_to_location.get(x, (None, None))[0])
joined_pandas_df['longitude'] = joined_pandas_df['customer_zip'].map(lambda x: zip_to_location.get(x, (None, None))[1])

# Filter out any rows with missing lat/lon
filtered_df = joined_pandas_df.dropna(subset=['latitude', 'longitude'])

#filtered_df['total_purchased_quantity'] = filtered_df['total_purchased_quantity'].astype(str)

#filtered_df['view_text'] = 'Total Stocks Purchased' + ' - ' + filtered_df['total_purchased_quantity'] + ',\nSELL COUNT' + ' - ' + filtered_df['sell_count'] + ' , \nTOTAL VALUE ' + ' - $' + filtered_df['total_value']


fig = px.scatter_geo(filtered_df,
                     lat='latitude',
                     lon='longitude',
                     size='total_purchased_quantity',  
                     color='trade_date',
                     text='customer_state',  
                     title=f'Stock Purchases by State based on date',
                     scope='usa')

fig.update_traces(marker=dict(opacity=0.6, line=dict(width=0.5, color='DarkSlateGrey')))
fig.update_layout(
    width=1200,   # Set the width (in pixels)
    height=800    # Set the height (in pixels)
)

fig.show()

```
